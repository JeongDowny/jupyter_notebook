```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
file = ('C:\\Users\\daun5\\jupyter notebook\\data\\manhattan.csv')
```

```python
df = pd.read_csv(file)
x = df[['bedrooms',
        'bathrooms',
        'size_sqft',
        'min_to_subway',
        'floor',
        'building_age_yrs',
        'no_fee',
        'has_roofdeck',
        'has_washer_dryer',
        'has_doorman',
        'has_elevator',
        'has_dishwasher',
        'has_patio',
        'has_gym']]
y = df[['rent']]

x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.8, test_size=0.2)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
my_apartment =[[0, 1, 480, 9, 2, 17, 1, 1, 0, 0, 1, 1, 0, 1]]
my_predict = lr.predict(my_apartment)
my_predict
```

```python
y_predict = lr.predict(x_test)
y_predict
```

```python
plt.scatter(y_test, y_predict, alpha=0.4)
plt.xlabel("Actual Rent")
plt.ylabel("Predicted Rent")
plt.show()
```

```python
lr.score(x_test, y_test)
```

```python
r2 = 1 - ((y_test - y_predict)**2).sum() / ((y_test - y_test.mean())**2).sum()
print(r2)
```
