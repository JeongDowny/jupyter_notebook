```python
import csv
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
stock_list = []
    
for page in range(1, 37):
    url = f'https://finance.naver.com/sise/sise_market_sum.naver?&page={page}'
    req = requests.get(url)
    soup = BeautifulSoup(req.text, 'html.parser')
    
    menu = soup.find("thead").find_all("th")
    menu_head = [head.get_text() for head in menu]
    menu_head = menu_head[1:-1]
    
    stocks_list = soup.find("table", attrs={"class": "type_2"}).find("tbody").find_all("tr")
    stocks = [stock.get_text().split() for stock in stocks_list]
    
    for i in stocks:
        if(i != []):
            i = i[1:]
            stock_list.append(i)
```

```python
df = pd.DataFrame()
for head in menu_head:
    df[head] = 0
```

```python
j = 0
for i in stock_list:
    l = len(i)

    if l == 11:
        df.loc[j] = i
    elif l == 12:
        i[1] = i[0] + ' ' + i[1]
        del i[0]
        df.loc[j] = i
    elif l == 13:
        i[2] = i[0] + ' ' + i[1] + ' ' + i[2]
        del i[0:2]
        df.loc[j] = i
    elif l == 14:
        i[3] = i[0] + ' ' + i[1] + ' ' + i[2] + ' ' + i[3]
        del i[0:3]
        df.loc[j] = i
    elif l == 15:
        i[4] = i[0] + ' ' + i[1] + ' ' + i[2] + ' ' + i[3] + ' ' + i[4]
        del i[0:4]
        df.loc[j] = i
    elif l == 16:
        i[5] = i[0] + ' ' + i[1] + ' ' + i[2] + ' ' + i[3] + ' ' + i[4] + ' ' + i[5]
        del i[0:5]
        df.loc[j] = i
    elif l == 17:
        i[6] = i[0] + ' ' + i[1] + ' ' + i[2] + ' ' + i[3] + ' ' + i[4] + ' ' + i[5] + ' ' + i[6]
        del i[0:6]
        df.loc[j] = i
    else:
        print(l)
    '''
    elif l > 11:
        print(i)
        for z in range(l-10):
            i[l-11] += i[z]
    '''
    j += 1
```
